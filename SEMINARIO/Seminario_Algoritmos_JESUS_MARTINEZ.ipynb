*¿Es un problema de maximización o minimización?*

El problema no se trata de un problema de maximización o minimización en sí mismo. Más bien, es un problema de encontrar una combinación de cifras y operadores que den como resultado una cantidad específica, sin repetir ninguno de ellos y utilizando las cuatro operaciones básicas.

En este caso, no hay una función objetivo que se busque maximizar o minimizar. En lugar de eso, el objetivo es encontrar una solución que cumpla con todas las restricciones y dé como resultado la cantidad deseada.

Por lo que este es un problema de fuerza bruta

**Diseña un algoritmo para resolver el problema por fuerza bruta**

```
import itertools

def Find_Combinations_BF(target):
    digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    operators = ['+', '-', '*', '/']
    solutions = []

    # Generar todas las posibles combinaciones de números y operadores
    for perm_digits in itertools.permutations(digits, 5):
        for perm_operators in itertools.permutations(operators, 4):
            # Construir la expresión
            expression = f"{perm_digits[0]} {perm_operators[0]} {perm_digits[1]} {perm_operators[1]} {perm_digits[2]} {perm_operators[2]} {perm_digits[3]} {perm_operators[3]} {perm_digits[4]}"
            try:
                # Evaluar la expresión
                result = eval(expression)
                if result == target:
                    solutions.append(expression)
            except ZeroDivisionError:
                pass # Ignorar divisiones por cero

    return solutions

target = 4
print("Numero: ",target)
solutions = Find_Combinations_BF(target)

 # Mostrar las soluciones encontradas
if len(solutions) > 0:
        print("Primera solucion: ",solutions[0])
        print(eval(solutions[0]))
        print(f"Se encontraron {len(solutions)} soluciones")
        # for solution in solutions:
        #     print(solution)
else:
        print("No se encontró ninguna solución.")
```

**Calcula la complejidad del algoritmo por fuerza bruta**

Respuesta

Dado que estamos trabajando con 9 números y 4 operadores, la cantidad total de combinaciones posibles es
9! (factorial de 9) para los números y 4! para los operadores. La complejidad del algoritmo podria ser calculada por el número de permutaciones de números. La fórmula general para la complejidad sería: **O(n⋅m)**

Donde:

n es el número de permutaciones de números (9!),
m es el número de permutaciones de operadores (4!).

Dado que 9! y 4! son relativamente pequeños (362,880 y 24, respectivamente), la complejidad total del algoritmo es manejable y podría considerarse como O(1) en términos prácticos.
